```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
X, y = make_blobs(n_samples=1000,
                  centers=4,
                  cluster_std=0.5,
                  random_state=0)

plt.figure(figsize=(12, 12))
plt.scatter(X[:,0], X[:,1])
plt.title("Original Data Set")
plt.show()
```

```python
kmeans = KMeans(n_clusters=2,
               random_state=0)

y_pred = kmeans.fit_predict(X)

plt.figure(figsize=(12, 12))
plt.scatter(X[:,0], X[:,1], c=y_pred)
plt.title("2 Clusters")
plt.show()
```

```python
kmeans = KMeans(n_clusters=4,
                random_state=0)
y_pred = kmeans.fit_predict(X)

plt.figure(figsize=(12, 12))
plt.scatter(X[:,0], X[:,1], c=y_pred)
plt.title("4 Clusters")
plt.show()
```

```python
X, y = make_blobs(n_samples=1000,
                  centers=3,
                  cluster_std=[1.0, 2.5, 0.5],
                  random_state=0)

plt.figure(figsize=(12, 12))
plt.scatter(X[:,0], X[:,1])
plt.title("Unequal Variances Data Set")
plt.show()
```

```python
kmeans = KMeans(n_clusters=3,
                random_state=0)
y_pred = kmeans.fit_predict(X)
```

```python
plt.figure(figsize=(12, 12))
plt.scatter(X[:,0], X[:,1], c=y_pred)
plt.title("3 Clusters")
plt.show()
```

# Gaussian Mixture Models

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.mixture import GaussianMixture
```

```python
X, y = make_blobs(n_samples=1000,
                  centers=4,
                  cluster_std=0.5,
                  random_state=0)
```

```python
GaussianM = GaussianMixture(n_components=4)

y_fit = GaussianM.fit(X)

y_pred = GaussianM.predict(X)
```

```python
GaussianM.aic(X)
```

```python
GaussianM.bic(X)
```

```python
plt.figure(figsize=(12, 12))
plt.scatter(X[:,0], X[:,1], c=y_pred)
plt.title("3 Clusters")
plt.show()
```
