## Revisiting the digits dataset

Recall our digits dataset comprised of 8x8 images as the input, with digits 0-9 as the target.

We can view each data point input as a vector of 64 components. Using PCA and t-SNE to project these data points down to 2 dimensions, we can plot the data points on a 2D graph and see if there is an obvious relationship between the output digits (which are indicated using colours) and the location of the data points on the graph.

```python
from sklearn.datasets import load_digits
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE

import matplotlib.pyplot as plt
```

```python
digits = load_digits()
print("Size of the digits dataset is: ", digits.data.shape)
print("The first digit in the dataset is", digits.target[0])
print("The first digit in the dataset has pixel intensities\n", digits.data[0])
```

```python
pca = PCA(n_components=2)
reduced_digits = pca.fit_transform(digits.data)
```

```python
plt.figure(figsize=(12, 12))
plt.scatter(reduced_digits[:,0], reduced_digits[:,1], c=digits.target)
plt.title("Handwritten digits using PCA")
plt.show()
```

```python
tsne = TSNE(n_components=2,
            init="pca")
reduced_digits = tsne.fit_transform(digits.data)
```

```python
plt.figure(figsize=(12, 12))
plt.scatter(reduced_digits[:,0], reduced_digits[:,1], c=digits.target)
plt.title("Handwritten digits using t-SNE")
plt.show()
```
