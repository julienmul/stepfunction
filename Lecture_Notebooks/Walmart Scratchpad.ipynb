```python
def reverser(x):
    reversed_list = []
    for ind in range(len(x)):
        reversed_list.append(x[-(ind+1)])
        
    return reversed_list

reverser([1,2,3,4])
```

```python
walmart_df = pd.read_csv('walmart_train.csv').head(30000)
```

```python
walmart_df.head()
```

```python
department_types = list(set(walmart_df.DepartmentDescription) - {np.nan})
for dept in department_types:
    walmart_df[dept] = 0
```

```python
for index, row in walmart_df.iterrows():
    col_name = row['DepartmentDescription']
    if not pd.isnull(col_name):
        walmart_df.loc[index, col_name] += walmart_df.loc[index, 'ScanCount']
```

```python
walmart_df.head()
```

```python
walmart_by_visit_num_df = walmart_df.groupby('VisitNumber')
function_map = {dept: np.sum for dept in department_types}
function_map['Weekday'] = pd.Series.mode
function_map['TripType'] = pd.Series.mode
walmart_by_visit_num_df = walmart_by_visit_num_df.agg(function_map)
```
